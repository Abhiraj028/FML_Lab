```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import spearmanr
data = pd.DataFrame({
    "source": ["A", "A", "B", "B", "C", "C", "D", "E", "F", "F", "G", "H", "H", "I", "J"],
    "target": ["B", "C", "C", "D", "A", "E", "C", "F", "C", "G", "C", "I", "C", "J", "C"],
    "weight": [1, 2, 1, 3, 1, 1, 2, 1, 4, 1, 1, 1, 2, 1, 1]
})
G = nx.from_pandas_edgelist(data, "source", "target", "weight", create_using=nx.DiGraph)


pr_global = nx.pagerank(G, alpha=0.5, weight="weight")   # change alpha here or remove it for default
print("Global Weighted PageRank")
for k, v in sorted(pr_global.items(), key= lambda x:x[1], reverse=True)[:5]:
    print(f"  {k}: {v:.4f}")


topic_nodes = {"C", "F"}
p_vec = {n: (1.0 / len(topic_nodes) if n in topic_nodes else 0.0) for n in G.nodes}
pr_topic = nx.pagerank(G, alpha = 0.5, weight="weight", personalization=p_vec, dangling=p_vec) # change alpha here or remove it for default
print(f"\nTopic-biased PageRank toward {topic_nodes} (top 5):")
for k, v in sorted(pr_global.items(), key= lambda x:x[1], reverse=True)[:5]:
    print(f"  {k}: {v:.4f}")


print("\nΔ (topic - global) for nodes (top shifts):")
delta = {n: pr_topic[n] - pr_global[n] for n in G.nodes}
for k,v in delta.items():
    print(f"{k}: {v:.4f}")

corr, _ = spearmanr(list(pr_global.values()), list(pr_topic.values()))
print(f"\nSpearman Rank Correlation (topic vs global): {corr:.4f}\n")

pos = nx.spring_layout(G, seed = 42) 
nx.draw_networkx(G, pos)
plt.title("Network Graph")
plt.show()
```
