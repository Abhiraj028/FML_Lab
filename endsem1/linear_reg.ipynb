```python
import numpy as np
from sklearn.datasets import fetch_california_housing, load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, explained_variance_score
import matplotlib.pyplot as plt

# Load dataset (vary dataset here)
data = fetch_california_housing()
# data = load_diabetes()

X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.3, random_state=42)

# Regularization: Ridge or Lasso (vary regularization here)
model = Ridge(alpha=1.0)  # vary alpha here: 0.1, 1.0, 10.0, etc.
# model = Lasso(alpha=1.0)  # vary alpha here: 0.1, 1.0, 10.0, etc.

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
evs = explained_variance_score(y_test, y_pred)

print(f"=== Regularized Linear Regression ===")
print(f"MSE: {mse:.3f}, MAE: {mae:.3f}, RMSE: {rmse:.3f}")
print(f"R2: {r2:.3f}, EVS: {evs:.3f}")

plt.scatter(y_test, y_pred)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'r--')
plt.title("Scatter Plot")
plt.show()
```
